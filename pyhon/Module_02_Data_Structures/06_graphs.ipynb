## 1. Graph Representations

### Adjacency List (Commonly used in interviews)

```python
class Graph:
    def __init__(self):
        self.adj_list = {}
        
    def add_edge(self, u, v, bidirectional=True):
        if u not in self.adj_list: self.adj_list[u] = []
        if v not in self.adj_list: self.adj_list[v] = []
        
        self.adj_list[u].append(v)
        if bidirectional:
            self.adj_list[v].append(u)

g = Graph()
g.add_edge(0, 1)
g.add_edge(0, 2)
print(f"Adjacency List: {g.adj_list}")
```

## 2. Graph Traversals

### DFS (Depth-First Search)

```python
def dfs(adj_list, start, visited=None):
    if visited is None: visited = set()
    visited.add(start)
    print(start, end=" ")
    
    for neighbor in adj_list.get(start, []):
        if neighbor not in visited:
            dfs(adj_list, neighbor, visited)

print("DFS traversal:")
dfs(g.adj_list, 0)
```

### BFS (Breadth-First Search)

```python
from collections import deque

def bfs(adj_list, start):
    visited = {start}
    queue = deque([start])
    
    while queue:
        u = queue.popleft()
        print(u, end=" ")
        
        for v in adj_list.get(u, []):
            if v not in visited:
                visited.add(v)
                queue.append(v)

print("\nBFS traversal:")
bfs(g.adj_list, 0)
```
