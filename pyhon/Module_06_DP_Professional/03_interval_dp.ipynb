## 1. Matrix Chain Multiplication (MCM)

**Problem**: Given a sequence of matrices, find the most efficient way to multiply these matrices together. The problem is not actually to perform the multiplications, but merely to decide in which order to perform the multiplications.

```python
def matrix_chain_order(p):
    n = len(p) - 1
    # dp[i][j] = min multiplications to multiply matrices from i to j
    dp = [[0] * (n + 1) for _ in range(n + 1)]
    
    # l is chain length
    for l in range(2, n + 1):
        for i in range(1, n - l + 2):
            j = i + l - 1
            dp[i][j] = float('inf')
            for k in range(i, j):
                # Split at k
                q = dp[i][k] + dp[k+1][j] + p[i-1]*p[k]*p[j]
                if q < dp[i][j]:
                    dp[i][j] = q
                    
    return dp[1][n]

p = [10, 30, 5, 60]
print(f"MCM Min Multiplications: {matrix_chain_order(p)}") # 4500
```

## 2. Burst Balloons

**Problem**: You are given `n` balloons, indexed from 0 to `n - 1`. Each balloon is painted with a number on it represented by an array `nums`. You are asked to burst all the balloons. If you burst balloon `i`, you will get `nums[left] * nums[i] * nums[right]` coins.

```python
def max_coins(nums):
    nums = [1] + nums + [1]
    n = len(nums)
    dp = [[0] * n for _ in range(n)]
    
    for length in range(2, n):
        for left in range(n - length):
            right = left + length
            for i in range(left + 1, right):
                # i is the LAST balloon to burst in the interval (left, right)
                dp[left][right] = max(dp[left][right], 
                                      nums[left] * nums[i] * nums[right] + 
                                      dp[left][i] + dp[i][right])
                                      
    return dp[0][n-1]

print(f"Max Coins [3,1,5,8]: {max_coins([3,1,5,8])}") # 167
```
