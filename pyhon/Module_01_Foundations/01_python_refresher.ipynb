## 1. Lists - The Most Important Data Structure

Lists are dynamic arrays in Python. Understanding list operations is crucial for DSA.

````python
# Creating lists
numbers = [1, 2, 3, 4, 5]
empty_list = []
mixed = [1, "hello", 3.14, True]

# Common operations
numbers.append(6)  # O(1) - Add to end
numbers.insert(0, 0)  # O(n) - Insert at index
numbers.pop()  # O(1) - Remove from end
numbers.pop(0)  # O(n) - Remove from beginning

# Slicing - Very important for DSA!
arr = [0, 1, 2, 3, 4, 5]
print(arr[1:4])  # [1, 2, 3] - from index 1 to 3
print(arr[:3])  # [0, 1, 2] - first 3 elements
print(arr[3:])  # [3, 4, 5] - from index 3 to end
print(arr[::2])  # [0, 2, 4] - every 2nd element
print(arr[::-1])  # [5, 4, 3, 2, 1, 0] - reverse
````

### List Comprehensions - Write Pythonic Code

````python
# Traditional way
squares = []
for i in range(10):
    squares.append(i**2)

# List comprehension (BETTER!)
squares = [i**2 for i in range(10)]
print(squares)

# With condition
even_squares = [i**2 for i in range(10) if i % 2 == 0]
print(even_squares)

# Nested comprehension (for 2D arrays)
matrix = [[i + j for j in range(3)] for i in range(3)]
print(matrix)
# [[0, 1, 2], [1, 2, 3], [2, 3, 4]]
````

## 5. Lambda Functions & Functional Programming

````python
# Lambda functions - anonymous functions
square = lambda x: x**2
print(square(5))  # 25

# Useful with sorted(), map(), filter()
points = [(3, 2), (1, 5), (2, 1)]

# Sort by x coordinate
sorted_by_x = sorted(points, key=lambda p: p[0])
print(sorted_by_x)  # [(1, 5), (2, 1), (3, 2)]

# Sort by distance from origin
sorted_by_dist = sorted(points, key=lambda p: p[0] ** 2 + p[1] ** 2)
print(sorted_by_dist)  # [(2, 1), (3, 2), (1, 5)]

# map() - apply function to all elements
nums = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, nums))
print(squared)  # [1, 4, 9, 16, 25]

# filter() - keep elements that satisfy condition
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)  # [2, 4]
````

## Problem 1: Two Sum (Easy)
Given an array of integers and a target, return indices of two numbers that add up to target.

```
Input: nums = [2, 7, 11, 15], target = 9
Output: [0, 1]
Explanation: nums[0] + nums[1] = 2 + 7 = 9
```

````python
def two_sum(nums, target):
    """
    Try to solve this yourself first!
    Hint: Use a dictionary to store seen numbers and their indices
    """

    # YOUR CODE HERE
    seen = {}
    for i, num in enumerate(nums):
        if target - num in seen:
            return [seen[target - num], i]
        seen[num] = i
    pass


# Test
print(two_sum([2, 7, 11, 15], 9))  # [0, 1]
print(two_sum([3, 2, 4], 6))  # [1, 2]
````

````python
# SOLUTION (Don't peek until you've tried!)
def two_sum_solution(nums, target):
    seen = {}  # value -> index
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    return []


# Time: O(n), Space: O(n)
print(two_sum_solution([2, 7, 11, 15], 9))
````

## Problem 4: Product of Array Except Self (Medium)
Return an array where each element is the product of all elements except itself.

```
Input: [1,2,3,4]
Output: [24,12,8,6]
Explanation: 24=2*3*4, 12=1*3*4, 8=1*2*4, 6=1*2*3
```
**Constraint**: Don't use division!

````python
def product_except_self(nums):
    """
    Hint: Calculate left and right products separately
    """
    # YOUR CODE HERE
    n=len(nums)
    output = [1]*n
    
    left_product=1
    for i in range(n):
        output[i] = left_product
        left_product *= nums[i]
        
    right_product=1
    for i in range(n-1,-1,-1):
        output[i]*=right_product
        right_product *=nums[i]
    
    return output


print(product_except_self([1, 2, 3, 4]))  # [24,12,8,6]
````

````python
# SOLUTION
def product_except_self_solution(nums):
    n = len(nums)
    result = [1] * n

    # Left products
    left = 1
    for i in range(n):
        result[i] = left
        left *= nums[i]

    # Right products
    right = 1
    for i in range(n - 1, -1, -1):
        result[i] *= right
        right *= nums[i]

    return result


# Time: O(n), Space: O(1) excluding output
print(product_except_self_solution([1, 2, 3, 4]))
````

````python
# SOLUTION
def longest_consecutive_solution(nums):
    if not nums:
        return 0

    num_set = set(nums)
    max_length = 0

    for num in num_set:
        # Only start counting from the beginning of a sequence
        if num - 1 not in num_set:
            current = num
            length = 1

            while current + 1 in num_set:
                current += 1
                length += 1

            max_length = max(max_length, length)

    return max_length


# Time: O(n), Space: O(n)
print(longest_consecutive_solution([100, 4, 200, 1, 3, 2]))
````
