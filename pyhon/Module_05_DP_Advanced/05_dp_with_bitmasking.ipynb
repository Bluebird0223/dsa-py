## 1. What is Bitmasking?

In DP, we often need to keep track of a set of items (e.g., visited nodes). If the number of items is small (usually $N \le 20$), we can use a single integer's bits to represent the subset. 

### Bitwise Operations Cheat Sheet
- `(1 << i)`: Get i-th bit
- `mask | (1 << i)`: Set i-th bit
- `mask & (1 << i)`: Check if i-th bit is set
- `mask ^ (1 << i)`: Toggle i-th bit


## 2. Traveling Salesman Problem (TSP)

**Problem**: Given a set of cities and distance between every pair of cities, find the shortest possible route that visits every city exactly once and returns to the starting point.

```python
def tsp(graph):
    n = len(graph)
    # dp[mask][i] = min distance visiting cities in mask, ending at i
    # Full mask = (1 << n) - 1
    dp = [[float('inf')] * n for _ in range(1 << n)]
    
    # Base case: start at city 0
    dp[1][0] = 0
    
    for mask in range(1, 1 << n):
        for u in range(n):
            if mask & (1 << u):
                for v in range(n):
                    if not (mask & (1 << v)):
                        new_mask = mask | (1 << v)
                        dp[new_mask][v] = min(dp[new_mask][v], dp[mask][u] + graph[u][v])
                        
    # Return to city 0
    full_mask = (1 << n) - 1
    ans = min(dp[full_mask][i] + graph[i][0] for i in range(1, n))
    return ans

graph = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]
print(f"TSP Min Distance: {tsp(graph)}") # 80
```
