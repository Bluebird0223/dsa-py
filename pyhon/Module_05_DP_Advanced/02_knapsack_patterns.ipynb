## 1. 0/1 Knapsack (The Classic)

**Problem**: Given weights and values of `n` items, put these items in a knapsack of capacity `W` to get the maximum total value. You can either include an item or exclude it (0/1).

**Recurrence**: `dp[i][w] = max(val[i] + dp[i-1][w-wt[i]], dp[i-1][w])` (if `wt[i] <= w`)

```python
def knapsack_01(weights, values, W):
    n = len(weights)
    # dp[i][w] = max value with first i items and capacity w
    dp = [[0] * (W + 1) for _ in range(n + 1)]
    
    for i in range(1, n + 1):
        for w in range(1, W + 1):
            if weights[i-1] <= w:
                # Max of (Include current item, Exclude current item)
                dp[i][w] = max(values[i-1] + dp[i-1][w - weights[i-1]], dp[i-1][w])
            else:
                dp[i][w] = dp[i-1][w]
                
    return dp[n][W]

# Space Optimized: O(W) space
def knapsack_01_optimized(weights, values, W):
    dp = [0] * (W + 1)
    for i in range(len(weights)):
        # Iterate backwards to avoid using the same item twice
        for w in range(W, weights[i] - 1, -1):
            dp[w] = max(values[i] + dp[w - weights[i]], dp[w])
    return dp[W]

weights = [1, 2, 3]
values = [60, 100, 120]
W = 5
print(f"0/1 Knapsack Max Value: {knapsack_01(weights, values, W)}") # 220
```

## 2. Subset Sum Pattern

**Problem**: Given a set of non-negative integers and a value `sum`, determine if there is a subset of the given set with sum equal to given `sum`.

```python
def can_partition(nums, target):
    dp = [False] * (target + 1)
    dp[0] = True
    
    for num in nums:
        for i in range(target, num - 1, -1):
            if dp[i - num]:
                dp[i] = True
                
    return dp[target]

print(f"Subset Sum [1, 5, 11, 5], 11: {can_partition([1, 5, 11, 5], 11)}") # True
```
