## 1. Longest Increasing Subsequence (LIS)

**Problem**: Given an integer array `nums`, return the length of the longest strictly increasing subsequence.

**Standard DP (O(nÂ²))**: `dp[i] = 1 + max(dp[j])` for all `j < i` and `nums[j] < nums[i]`

```python
def length_of_lis(nums):
    if not nums: return 0
    dp = [1] * len(nums)
    
    for i in range(len(nums)):
        for j in range(i):
            if nums[i] > nums[j]:
                dp[i] = max(dp[i], dp[j] + 1)
                
    return max(dp)

print(f"LIS [10,9,2,5,3,7,101,18]: {length_of_lis([10,9,2,5,3,7,101,18])}") # 4
```

## 2. LIS Optimized (O(n log n))

Using Binary Search we can reduce the time complexity significantly. This is a common interview requirement for large inputs.

```python
import bisect

def length_of_lis_optimized(nums):
    sub = []
    for x in nums:
        if not sub or sub[-1] < x:
            sub.append(x)
        else:
            idx = bisect.bisect_left(sub, x)
            sub[idx] = x
    return len(sub)

print(f"LIS Optimized [0,1,0,3,2,3]: {length_of_lis_optimized([0,1,0,3,2,3])}") # 4
```
