## 1. The House Robber Pattern (Include/Exclude)

The House Robber problem is a classic introduction to the "Include/Exclude" pattern in DP. 

**Problem**: You are a professional robber planning to rob houses along a street. Each house has a certain amount of money stashed. The only constraint stopping you from robbing each of them is that adjacent houses have security systems connected and it will automatically contact the police if two adjacent houses were broken into on the same night.

**Goal**: Maximize the amount of money you can rob.

```python
def rob(nums):
    if not nums: return 0
    if len(nums) == 1: return nums[0]
    
    # dp[i] = max money up to house i
    dp = [0] * len(nums)
    
    # Base cases
    dp[0] = nums[0]
    dp[1] = max(nums[0], nums[1])
    
    for i in range(2, len(nums)):
        # Either rob current house (add dp[i-2]) OR skip current house (take dp[i-1])
        dp[i] = max(nums[i] + dp[i-2], dp[i-1])
        
    return dp[-1]

# Space Optimized: O(1) space
def rob_optimized(nums):
    if not nums: return 0
    prev2 = 0
    prev1 = 0
    
    for num in nums:
        current = max(num + prev2, prev1)
        prev2 = prev1
        prev1 = current
        
    return prev1

print(f"House Robber [1,2,3,1]: {rob([1,2,3,1])}") # 4
print(f"House Robber Optimized [2,7,9,3,1]: {rob_optimized([2,7,9,3,1])}") # 12
```

## 2. Min Cost Climbing Stairs

**Problem**: You are given an integer array `cost` where `cost[i]` is the cost of `i-th` step on a staircase. Once you pay the cost, you can either climb one or two steps. You can either start from the step with index 0, or index 1.

**Goal**: Minimum cost to reach the top of the floor.

```python
def min_cost_climbing_stairs(cost):
    n = len(cost)
    # dp[i] = min cost to reach step i
    dp = [0] * (n + 1)
    
    for i in range(2, n + 1):
        # To reach step i, we could come from i-1 or i-2
        dp[i] = min(dp[i-1] + cost[i-1], dp[i-2] + cost[i-2])
        
    return dp[n]

print(f"Min Cost [10, 15, 20]: {min_cost_climbing_stairs([10, 15, 20])}") # 15
```

## 3. Jump Game Variations

### Jump Game I: Can you reach the end?
While this can be solved with DP in O(nÂ²) or O(n), the Greedy approach is O(n) and O(1) space.

```python
def can_jump(nums):
    gas = 0
    for n in nums:
        if gas < 0: return False
        if n > gas:
            gas = n
        gas -= 1
    return True

print(f"Can Jump [2,3,1,1,4]: {can_jump([2,3,1,1,4])}") # True
print(f"Can Jump [3,2,1,0,4]: {can_jump([3,2,1,0,4])}") # False
```

### Jump Game II: Minimum Jumps

```python
def jump(nums):
    n = len(nums)
    if n <= 1: return 0
    
    # DP approach: dp[i] = min jumps to reach index i
    dp = [float('inf')] * n
    dp[0] = 0
    
    for i in range(n):
        for j in range(1, nums[i] + 1):
            if i + j < n:
                dp[i + j] = min(dp[i + j], dp[i] + 1)
                
    return dp[-1]

print(f"Min Jumps [2,3,1,1,4]: {jump([2,3,1,1,4])}") # 2
```
