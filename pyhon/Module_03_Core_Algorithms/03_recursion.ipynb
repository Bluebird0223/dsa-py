## 1. Recursion Fundamentals

A recursive function has two main parts:
1. **Base Case**: The condition where the recursion stops.
2. **Recursive Step**: The function calling itself with a simpler version of the problem.

```python
def factorial(n):
    # Base case
    if n <= 1: return 1
    # Recursive step
    return n * factorial(n - 1)

print(f"Factorial 5: {factorial(5)}") # 120
```

## 2. Backtracking Pattern

Backtracking is used to find all (or some) solutions to a problem by incrementally building candidates and abandoning them if they cannot lead to a valid solution.

```python
def subsets(nums):
    res = []
    
    def backtrack(start, current_path):
        # Every path we take is a valid subset
        res.append(current_path[:])
        
        for i in range(start, len(nums)):
            # Choose
            current_path.append(nums[i])
            # Explore
            backtrack(i + 1, current_path)
            # Un-choose (Backtrack)
            current_path.pop()
            
    backtrack(0, [])
    return res

print(f"Subsets [1, 2]: {subsets([1, 2])}") # [[], [1], [1, 2], [2]]
```
