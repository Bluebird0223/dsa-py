### Exercise 2: Celsius to Fahrenheit
Convert celsius temperature to fahrenheit.
Formula: `(celsius * 9/5) + 32`

```python
def celsius_to_fahrenheit(celsius):
    # TODO: Write your code here
    return (celsius * 9/5) + 32
    pass
print(celsius_to_fahrenheit(0))
```

```python
# --- Test Exercise 2 ---
try:
    assert celsius_to_fahrenheit(0) == 32
    assert celsius_to_fahrenheit(100) == 212
    print("✅ Exercise 2 Passed!")
except AssertionError:
    print("❌ Exercise 2 Failed.")
except Exception as e:
    print(f"❌ Error: {e}")
```

### Exercise 3: Is Even?
Return `True` if the number is even, `False` otherwise.
Hint: Use the modulo operator `%`.

```python
def is_even(number):
    # TODO: Write your code here
    return number % 2 == 0
    pass
print(is_even(2))
```

### Exercise 4: Check Sign
Return `"Positive"` if number is > 0.
Return `"Negative"` if number is < 0.
Return `"Zero"` if number is 0.

```python
def check_sign(number):
    # TODO: Write your code here
    if number > 0:
        return "Positive"
    elif number < 0:
        return "Negative"
    else:
        return "Zero"
    pass
print(check_sign(1))
```

```python
# --- Test Exercise 4 ---
try:
    assert check_sign(10) == "Positive"
    assert check_sign(-5) == "Negative"
    assert check_sign(0) == "Zero"
    print("✅ Exercise 4 Passed!")
except AssertionError:
    print("❌ Exercise 4 Failed.")
except Exception as e:
    print(f"❌ Error: {e}")
```

### Exercise 5: Max of Three
Return the largest of three numbers `a`, `b`, and `c`.
Try to use `if/elif/else` comparisons.

```python
def max_of_three(a, b, c):
    # TODO: Write your code here
    # print(a,b,c)
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c
    pass


print(max_of_three(1, 2, 3))
```
