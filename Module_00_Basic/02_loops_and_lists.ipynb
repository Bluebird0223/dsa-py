### Exercise 2: Sum List
Return the sum of all numbers in the given list.
Do **not** use the built-in `sum()` function, use a loop.

```python
def sum_list(numbers):
    # TODO: Write your code here
    sum=0
    for i in numbers:
        sum+=i
    return sum
    pass
```

```python
# --- Test Exercise 2 ---
try:
    assert sum_list([1, 2, 3, 4, 5]) == 15
    assert sum_list([-1, 1]) == 0
    assert sum_list([]) == 0
    print("✅ Exercise 2 (sum_list) passed!")
except AssertionError:
    print("❌ Exercise 2 (sum_list) failed.")
except Exception as e:
    print(f"❌ Error: {e}")
```

### Exercise 4: Find Maximum
Return the largest number in the list.
Do **not** use the built-in `max()` function.

```python
def find_maximum(numbers):
    # TODO: Write your code here
    maximum=numbers[0]
    for i in numbers[1:]:
        
        if maximum <= i:
            print(i,maximum)
            maximum=i
    return maximum
    pass
```

### Exercise 5: Filter Evens
Return a **NEW** list containing only the even numbers from the original list.
Example: `filter_evens([1, 2, 3, 4])` -> `[2, 4]`

```python
def filter_evens(numbers):
    # TODO: Write your code here
    even=[]
    for i in numbers:
        if i%2==0:
            even.append(i)
    return even
    pass
```

```python
# --- Test Exercise 5 ---
try:
    assert filter_evens([1, 2, 3, 4, 5, 6]) == [2, 4, 6]
    assert filter_evens([1, 3, 5]) == []
    print("✅ Exercise 5 (filter_evens) passed!")
except AssertionError:
    print("❌ Exercise 5 (filter_evens) failed.")
except Exception as e:
    print(f"❌ Error: {e}")
```
